```python
import numpy as np
import json
import pandas as pd
from shapely.wkt import dumps, loads
from shapely.geometry import Polygon
```

```python
df = open('/home/ayush/Documents/VisionBeyondLimits/Labels/mexico-earthquake_00000129_post_disaster.json')
data = json.load(df)
```

```python
for a in data:
    print(a)
```

```python
subtype =[]
points = []
for coordinate in data['features']['xy']:
    #print(coordinate['wkt'])
    subtype.append(coordinate['properties']['subtype'])
    points.append(coordinate['wkt'])
    #print("--")
```

```python
from shapely import wkt
p1=[]
for o in points:
    p1.append(wkt.loads(o))
```

```python
from matplotlib import image
import matplotlib.pyplot as plt
from shapely.geometry import Polygon as POLYGON
img = image.imread('/home/ayush/Documents/VisionBeyondLimits/Images/mexico-earthquake_00000129_post_disaster.png')
for k in range(len(p1)):
    if(subtype[k]=='no-damage'):
        x,y = p1[k].exterior.xy
        plt.fill(x,y,color="yellow")
    elif(subtype[k]=='un-classified'):
        plt.fill(x,y,color="blue")
    elif(subtype[k]=='minor-damage'):
        plt.fill(x,y,color="red")
    elif(subtype[k]=='major-damage'):
        plt.fill(x,y,color="green")
    elif(subtype[k]=='destroyed'):
        plt.fill(x,y,color="black")
    plt.imshow(img)
plt.show()
```
